```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, roc_auc_score
from sklearn.metrics import classification_report, precision_recall_fscore_support
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df= pd.read_csv("/home/jeffy/Downloads/ML/CICIoT2023/part-00000-363d1ba3-8ab5-4f96-bc25-4d5862db7cb9-c000.csv")
```

```python
df["label"].value_counts()
```

```python
df["label"] = np.where(df["label"] == "BenignTraffic", 0, 1)
```

```python
df["label"].value_counts()
```

```python
df["label"].hist(bins=2)
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["label"])
train_df
```

```python
test_df["label"].hist(bins=2)
```

```python
features=list(train_df.columns)
features.remove("label")
```

```python
scaler = StandardScaler()

train_df[features] = scaler.fit_transform(train_df[features])
test_df[features] = scaler.transform (test_df[features])
test_df
```

```python
X_train = train_df[features].values
y_train = train_df["label"].values

X_test = test_df[features].values
y_test = test_df["label"].values
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(max_iter =120000)
clf.fit(X_train, y_train.ravel())
```

```python
y_predict = clf.predict(X_test)
clf_score = clf.score(X_test,y_test)
clf_score
```

```python
print('Accuracy of LG:' + (str(clf_score)))
precision, recall, fscore, none = precision_recall_fscore_support(y_test, y_predict, average="weighted")

print('Precision of LG:' + (str(precision)))
print('Recall of LG:' + (str(recall)))
print('F1 Score of LG:' + (str(fscore)))
print(classification_report(y_test, y_predict))
```

```python
cm=confusion_matrix(y_test, y_predict)
f,ax= plt.subplots(figsize=(10,10))
sns.heatmap(cm, annot=True, linewidth=0.5, linecolor='red', fmt = '.0f',ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
```

Tree Classifier

```python
tree=DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)

tree_score = tree.score(X_test,y_test)
y_predict = tree.predict(X_test)

```

```python
print('Accuracy of Tree:' + (str(tree_score)))
precision, recall, fscore, none = precision_recall_fscore_support(y_test, y_predict, average="weighted")

print('Precision of Tree:' + (str(precision)))
print('Recall of Tree:' + (str(recall)))
print('F1 Score of Tree:' + (str(fscore)))
print(classification_report(y_test, y_predict))
```

```python
cm=confusion_matrix(y_test, y_predict)
f,ax= plt.subplots(figsize=(10,10))
sns.heatmap(cm, annot=True, linewidth=0.5, linecolor='red', fmt = '.0f',ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
```
